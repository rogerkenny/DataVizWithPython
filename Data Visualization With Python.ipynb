```python
cprimes = [7, 19, 37, 61, 127, 271, 331, 397, 547, 631, 919, 1657, 1801, 1951, 2269, 2437, 2791, 3169, 3571, 4219, 4447, 5167, 5419, 6211, 7057, 7351, 8269, 9241, 10267, 11719, 12097, 13267, 13669, 16651, 19441, 19927, 22447, 23497, 24571, 25117, 26227, 27361, 33391, 35317]

# turn = math.log2(22)
turn = (1 + math.sqrt(5)) / 2  #1.61803398875 #golden ratio
# turn = 1/5.5515255521414
# turn = math.e
# turn = math.pi
nodes = 200
scale = 0.1
r = [i for i in range(nodes)]
fx = [math.cos(i * turn * 2 * math.pi) * i * scale for i in r]
fy = [math.sin(i * turn * 2 * math.pi) * i * scale for i in r]
plt.figure(num=1, figsize=(8,8))
plt.plot(fx,fy, 'ko')
plt.ylim([-20.1,20.1])
plt.xlim([-20.1,20.1])
plt.show()
```
